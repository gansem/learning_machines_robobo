```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline  
import pandas as pd
import info

# path = f'../results/{info.task}/{info.user}/take_03/learning_progress.tsv'
# print(path)
# df = pd.read_csv(path, sep='\t')
scene = '6'
dumb_df = pd.read_csv('C:/Users/Hammock/Documents/VU_masters_AI/learning_machines/learning_machines_robobo/results/obstacle_avoidance/dumb_robobo/scene_'+scene+'/dumb_robobo_progress.tsv', sep='\t')
smart_df = pd.read_csv('C:/Users/Hammock/Documents/VU_masters_AI/learning_machines/learning_machines_robobo/results/obstacle_avoidance/smart_robobo_0300h/scene_'+scene+'/learning_progress.tsv', sep='\t')
smarter_df = pd.read_csv('C:/Users/Hammock/Documents/VU_masters_AI/learning_machines/learning_machines_robobo/results/obstacle_avoidance/smart_robobo_0650h/scene_'+scene+'/learning_progress.tsv', sep='\t')
```

```python
# rename index col
dumb_df = dumb_df.rename(columns={'Unnamed: 0': 'index'})
smart_df = smart_df.rename(columns={'Unnamed: 0': 'index'})
smarter_df = smarter_df.rename(columns={'Unnamed: 0': 'index'})

dfs = [dumb_df, smart_df, smarter_df]
```

```python
dumb_df.head()
```

```python
distances = []
for _df in dfs:
    acc_d = 0
    acc_distance = []
    for i, v in enumerate(_df['v_distance_reward']):
        acc_d += v
        acc_distance.append(acc_d)
    distances.append(acc_distance)

_df = pd.DataFrame({'index': dumb_df['index'], 'dumb_distance': distances[0], 'smart_distance': distances[1], 'smarter_distance': distances[2]})
_df.head()
```

```python

# group per episode and append to new df
# _df = pd.DataFrame()
# for k,v in df.groupby('episode_index'):
#     _df = _df.append({'episode_index': int(k), 'reward': v['reward'].sum(), 'v_measure_sensor_distance': v['v_measure_sensor_distance'].sum(), 'epsilon': v['epsilon'].min(), 'v_distance_reward': v['v_distance_reward'].sum()}, ignore_index=True)
    # print(v['reward'].sum())

    
# plot 
fig, ax = plt.subplots(figsize=(15,10))
melt = _df.melt(id_vars='index', value_vars=['dumb_distance', 'smart_distance', 'smarter_distance'])
sns.lineplot(data=melt, x='index', y='value', hue='variable')
plt.legend()
img_path = 'C:/Users/Hammock/Documents/VU_masters_AI/learning_machines/learning_machines_robobo/results/obstacle_avoidance/'
plt.savefig(img_path+'/learning_rate_lineplot_'+scene+'.png')
```
